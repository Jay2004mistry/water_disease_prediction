```python
df["Water Treatment Method"]=df["Water Treatment Method"].fillna(df["Water Source Type"].mode()[0])
```

```python

#encode the sreing value to numeric
from sklearn.preprocessing import LabelEncoder

le_country = LabelEncoder()
le_region  = LabelEncoder()
le_source  = LabelEncoder()
le_treat   = LabelEncoder()

df['Country']              = le_country.fit_transform(df['Country'])
df['Region']               = le_region.fit_transform(df['Region'])
df['Water Source Type']    = le_source.fit_transform(df['Water Source Type'])
df['Water Treatment Method'] = le_treat.fit_transform(df['Water Treatment Method'])

# now print the mappings
print("COUNTRY mapping:")
print(dict(zip(le_country.classes_, le_country.transform(le_country.classes_))))

print("\nREGION mapping:")
print(dict(zip(le_region.classes_, le_region.transform(le_region.classes_))))

print("\nWATER SOURCE mapping:")
print(dict(zip(le_source.classes_, le_source.transform(le_source.classes_))))

print("\nTREATMENT mapping:")
print(dict(zip(le_treat.classes_, le_treat.transform(le_treat.classes_))))
```

```python
#divide the data
# ---> x train is data that use to as input. already given to model
# ---> y train is use as labeled output already given to the model 
# ---> x test is use as test data which model see first time after train from this data it 
# predict(unseen data)
# ---> y test is actual output and (output produced by (x test) data will compare)
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(["Diarrheal Cases per 100,000 people","Cholera Cases per 100,000 people","Typhoid Cases per 100,000 people","Infant Mortality Rate (per 1,000 live births)"],axis=1)
y=df[["Diarrheal Cases per 100,000 people","Cholera Cases per 100,000 people","Typhoid Cases per 100,000 people","Infant Mortality Rate (per 1,000 live births)"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Train data",x_train.shape)
print("test data",x_test.shape)
```

```python
from sklearn.metrics import cohen_kappa_score,classification_report,accuracy_score,roc_auc_score,roc_curve,confusion_matrix
```

```python

#it is regression so function changed. classification function will diffrent
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

def evaluate_model(y_test, y_pred):
    names = ['Diarrheal', 'Cholera', 'Typhoid', 'Infant Mortality']
    
    for i, name in enumerate(names):
        r2   = r2_score(y_test.iloc[:, i], y_pred[:, i])
        mae  = mean_absolute_error(y_test.iloc[:, i], y_pred[:, i])
        rmse = np.sqrt(mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
        print(f"{name} => R²: {r2:.2f}  MAE: {mae:.2f}  RMSE: {rmse:.2f}")
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train_sc, y_train)
y_pred_lr = lr.predict(x_test_sc)

print("LINEAR REGRESSION")
evaluate_model(y_test, y_pred_lr)
```

```python
#  Decision Tree
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(max_depth=6, random_state=42)
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)

print("DECISION TREE")
evaluate_model(y_test, y_pred_dt)
```

```python
#  Random Forest
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)

print("RANDOM FOREST")
evaluate_model(y_test, y_pred_rf)
```

```python
#  SVR
from sklearn.svm import SVR
from sklearn.multioutput import MultiOutputRegressor

svr = MultiOutputRegressor(SVR(kernel='rbf'))
svr.fit(x_train_sc, y_train)
y_pred_svr = svr.predict(x_test_sc)

print("SVR")
evaluate_model(y_test, y_pred_svr)
```

```python
# KNN
from sklearn.neighbors import KNeighborsRegressor

knn = MultiOutputRegressor(KNeighborsRegressor(n_neighbors=5))
knn.fit(x_train_sc, y_train)
y_pred_knn = knn.predict(x_test_sc)

print("KNN")
evaluate_model(y_test, y_pred_knn)
```

```python
#save the best model(Linear Regression is best here)
#pickle is python tools help to save the object  to file.
#here we need to save model object into file
import pickle

#save the linear regression model
#with open:- open file
# Linear_reg_model_for_disease_water.pkl:- file name
# wb:- Write in binary mode (for saving)
#as f:- indicate this file as 'f'
# pickle.dump(lr,f):- Saves the lr model into file f
with open('Linear_reg_model_for_disease_water.pkl','wb') as f:
    pickle.dump(lr,f)


#Save the scaler. used to normalized input data. reduce the complexity
with open('scaler.pkl','wb') as f:
    pickle.dump(scaler,f)

print("Model save")
```
